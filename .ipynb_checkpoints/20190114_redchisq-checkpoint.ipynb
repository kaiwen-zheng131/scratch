```python
def reduced_chi_squared(ydata, ymod, n_param=2,sd=None):
    chisq = np.sum((np.real(ydata) - np.real(ymod))**2/np.real(sd)**2) + np.sum((np.imag(ydata) - np.imag(ymod))**2/np.imag(sd)**2)
    
    chisq2 = np.sum((np.real(ydata) - np.real(ymod))**2/(2*np.real(sd))**2) + np.sum((np.imag(ydata) - np.imag(ymod))**2/(2*np.imag(sd))**2)

    nu=ydata.size-n_param
    red_chisq = chisq/nu 
    red_chisq2 = chisq2/nu
    return chisq, red_chisq, red_chisq2
```

```python
def residuals(ydata, ymod):
    return ydata-ymod
```

```python
s21_complex = np.vectorize(complex)(real_res, imag_res)
```

```python
resids = residuals(s21_complex, best_fit)
```

```python
redchisq = reduced_chi_squared(s21_complex, best_fit, n_param=9, sd=np.vectorize(complex)(std(resids.real), std(resids.imag)))
```
