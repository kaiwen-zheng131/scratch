```python
figure()
plot(freq, 20*np.log10(np.sqrt(real(response)**2 + imag(response)**2)))
```

```python
import resonator_model
```

```python
def get_qi(Q, Q_e_real):
    return (Q**-1 - Q_e_real**-1)**-1

def get_br(Q, f_0):
    return f_0*(Q)**-1
```

```python
plt.figure(figsize=(14,6))

real_res = np.array(np.real(response))
imag_res = np.array(np.imag(response))
freqs_res = freq*1e6

#fit the model
result = resonator_model.full_fit(freqs_res, real_res, imag_res)
f_0 = result.best_values['f_0']*1e-9
Q = result.best_values['Q']
Qc = result.best_values['Q_e_real']
Qi = get_qi(result.best_values['Q'], result.best_values['Q_e_real'])
br = get_br(result.best_values['Q'], result.best_values['f_0'])

#lets make a fine version of the model to plot
f_fine = arange(min(freqs_res), max(freqs_res), 100)
fine_s21 = resonator_model.fine_s21_model(f_fine, result.best_values)

#plot it
plt.subplot(121)
plt.plot(real_res, imag_res, '+', color = 'k', label = '$\mathrm{freq.\ sweep\ data}$')
plt.plot(fine_s21.real, fine_s21.imag, color = 'g', label = '$\mathrm{S_{21}\ model}$')
plt.xlabel('$\mathrm{I}$')
plt.ylabel('$\mathrm{Q}$')
plt.text(min(fine_s21.real), min(fine_s21.imag),'$\mathrm{Q_i: %.2f}$ \n $\mathrm{Q_c: %.2f}$ \n $\mathrm{Q: %.2f}$ \n $\mathrm{f_0: %.2f~GHz}$ \n $\mathrm{b_r: %.2f~Hz}$' %(Qi, Qc, Q, f_0, br), size = 16, color = 'k')
plt.legend(loc = 4)
plt.tight_layout()


plt.subplot(122)
plt.plot(freqs_res*1e-9, 20*np.log10(np.sqrt(real_res**2 + imag_res**2)), '+', color = 'k', label = '$\mathrm{data}$')
plt.plot(f_fine*1e-9, 20*np.log10(np.sqrt(fine_s21.real**2 + fine_s21.imag**2)), color = 'g', label = '$\mathrm{model}$')
plt.xlabel('$\mathrm{GHz}$')
plt.ylabel('$\mathrm{S_{21}~[dB]}$')
plt.legend(loc = 4)
plt.tight_layout()

tight_layout()
```
