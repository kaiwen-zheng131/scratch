```python
from lmfit import Model, Parameters, Minimizer
```

```python
def noise_model_with_alpha(f, white_noise, f_knee, alpha, f_c, amp):
    return ((white_noise)*(1 + (f_knee/f)**alpha)* (1/(1 + (f/f_c)**2))+amp)
```

```python
my_model = noise_model_with_alpha(f=fr, white_noise=3000, f_c = 50, f_knee=0.2, alpha = 1.5, amp = 100)
```

```python
loglog(fr, pxx, alpha = 0.6)
loglog(fr, my_model)
```

```python
gmod = Model(noise_model_with_alpha)
```

```python
result = gmod.fit(pxx,f =fr,white_noise=3000, f_c = 50, f_knee=0.2, alpha = 1.5)
```
